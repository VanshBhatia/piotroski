```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt


sensex=pd.read_csv("C:/Users/tradinglab11/Desktop/Vaibhav/Final Python Code/Sensex.txt",sep='|')
```

```python
sensex
```

```python
v=sensex.co_stkdate
y=v//10000
md=v%10000
d=md%100
m=md//100


sensex['Month']=m
sensex['year']=y
sensex['date']=d


sensex_new=sensex[sensex['Month']==8]

S=[]
for j in range(2005,2017):
    h=sensex_new[sensex_new['year']==j]
    k=min(h['date'])
    S.append(h[h['date']==k])
```

```python
S[2]
```

```python
T=[]
for i in range(2,len(S)-1):
     z=S[i].merge(S[i+1],how='left', left_on='company_name', right_on='company_name')
     z['bse_closing_price_y'].fillna(0,inplace=True)
     z['per_return']=(-z['bse_closing_price_y']+z['bse_closing_price_x'])/z['bse_closing_price_x']
     T.append(z)
```

```python
len(T)
```

```python
T[0]
```

```python
q3=[]
for k in range(0,len(T)):
    q3.append(T[k]['per_return'].mean())
```

```python
Sensex_ret = pd.DataFrame({'Avg': q3 }) 
```
