```python
df_0=pd.read_csv("C:/Users/tradinglab11/Desktop/Vaibhav/Final Python Code/returns.txt",sep='|')
df_1=df_0.dropna()

v=df_1.co_stkdate
y=v//10000
md=v%10000
d=md%100
m=md//100


df_1['Month']=m
df_1['year']=y
df_1['date']=d


df_new_1=df_1[df_1['Month']==8]

R=[]
for j in range(2005,2017):
    h=df_new_1[df_new_1['year']==j]
    k=min(h['date'])
    R.append(h[h['date']==k])
```

```python
Q=[]
P=[]
for i in range(0,len(F_Final)):
    z=F_Final[i].merge(R[i+2],how='left', left_on='Ca_company_name_x', right_on='company_name')
    z=z.dropna()
    z=z[z['bse_market_cap']>=1000]
    P.append(z)
    Q.append(P[i][P[i]['bse_pb']<P[i]['bse_pb'].quantile(.2)].sort(['F_Score']))
```

```python
I=[]
for i in range(0,len(Q)):
     z=Q[i].merge(R[i+3],how='left', left_on='Ca_company_name_x', right_on='company_name')
     z['bse_closing_price_y'].fillna(0,inplace=True)
     z['per_return']=(z['bse_closing_price_y']-z['bse_closing_price_x'])/z['bse_closing_price_x']
     del z['co_code_x']
     del z['company_name_x']
     del z['co_code_y']
     del z['company_name_y']
     z.to_csv(str(2007+i)+'.csv')
     I.append(z)
```

```python
J=[]
for i in range(0,len(Q)-1):
     z=I[i].merge(R[i+4],how='left', left_on='Ca_company_name_x', right_on='company_name')
     z['bse_closing_price'].fillna(0,inplace=True)
     z['per_return']=(z['bse_closing_price']-z['bse_closing_price_x'])/z['bse_closing_price_x']
     J.append(z)
```
